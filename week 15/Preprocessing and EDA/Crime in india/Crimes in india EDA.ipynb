```python
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame containing the crime data

# Group data by 'STATE/UT' and calculate total IPC crimes for each state
statewise_total_crimes = data.groupby('STATE/UT')['TOTAL IPC CRIMES'].sum().head(15)

# Sort states by total crimes in descending order
statewise_total_crimes = statewise_total_crimes.sort_values(ascending=False)

# Plot total IPC crimes by state
plt.figure(figsize=(12, 6))
statewise_total_crimes.plot(kind='bar', color='skyblue')
plt.title('Total IPC Crimes by State')
plt.xlabel('State')
plt.ylabel('Total IPC Crimes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Sum up the total number of crimes across all years for each crime type
crime_type_totals = data.drop(['STATE/UT', 'DISTRICT', 'YEAR'], axis=1).sum()

# Sort crime types by total crimes in descending order
crime_type_totals = crime_type_totals.sort_values(ascending=False)

# Plot the distribution of crime types
plt.figure(figsize=(12, 8))
crime_type_totals.plot(kind='barh', color='skyblue')
plt.title('Distribution of Crime Types')
plt.xlabel('Total Number of Crimes')
plt.ylabel('Crime Type')
plt.tight_layout()
plt.show()
```
