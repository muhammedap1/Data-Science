```python
import seaborn as sns
import matplotlib.pyplot as plt

# Creating the correlation matrix
correlation_matrix = data.corr(numeric_only=True)

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Scatter plot between "Age" and "Fare"
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', data=data)
plt.title('Scatter Plot: Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

# Bar plot showing the distribution of passenger classes ("Pclass")
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', data=data)
plt.title('Bar Plot: Distribution of Passenger Classes')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()

# Histogram of ages
plt.figure(figsize=(8, 6))
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Histogram: Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Box plot comparing the distribution of "Age" across different passenger classes
plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Age', data=data)
plt.title('Box Plot: Distribution of Age across Passenger Classes')
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.show()
```
