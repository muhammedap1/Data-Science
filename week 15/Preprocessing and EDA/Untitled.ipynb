```python
feature_with_na=[feature for feature in df.columns if df[feature].isnull().sum()>1]
```

```python
for feature in feature_with_na:
    print(feature,np.round(df[feature].isnull().mean(),5),'% of missing values')
```

```python
for feature in feature_with_na:
    data=df.copy()
    data[feature]=np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
    
```

```python
numerical_features=[feature for feature in df.columns if df[feature].dtype !='O']
print('number of numerical variables in the data of housing data:', len(numerical_features))
```

```python
df[numerical_features].head(3)
```

```python
year_feature=[feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
```

```python
year_feature
```

```python
for feature in year_feature:
    print(feature,df[feature].unique())
```

```python
discrete_features=[feature for feature in numerical_features if len(df[feature].unique())<25 and feature not in year_feature+['id'] ]
```

```python
len(discrete_features)
```

```python
for feature in discrete_features:
    data=df.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.show()
```

```python
continues_feature=[feature for feature in numerical_features if feature not in year_feature+['Id']+discrete_features]
```

```python
len(continues_feature)
```

```python
for feature in continues_feature:
    data=df.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('count')
    plt.show()
```

```python
for feature in continues_feature:
    data=df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

```python
def chage_cat(df,feature_with_na):
    data=df.copy()
    data[feature_with_na]=data[feature_with_na].fillna('Missing')
    return data

df=chage_cat(df,feature_with_na)
df[feature_with_na].isnull().sum()
```
