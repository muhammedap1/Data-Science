```python
import pandas as pd
```

```python
df=pd.read_csv('loan_data.csv')
```

```python
df
```

```python
nums_cols=df.select_dtypes(include=['float64','int64']).columns.tolist()
```

```python
Q1=df[nums_cols].quantile(0.25)
```

```python
Q3=df[nums_cols].quantile(0.75)
```

```python
IQR=Q3-Q1
```

```python
outliers=((df[nums_cols] < (Q1-1.5*IQR))|(df[nums_cols]>(Q3+1.5*IQR))).any(axis=1)
```

```python
df_clean=df[~outliers]
```

```python
df.shape
```

```python
df_clean.shape
```

```python
df_clean
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df_clean.drop(['not.fully.paid'],axis=1)
```

```python
y=df_clean['not.fully.paid']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
x_train=scalar.fit_transform(x_train)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df_clean['purpose']=le.fit_transform(df_clean['purpose'])
```

```python
df_clean
```

```python
x_train
```

```python
x_train=scalar.fit_transform(x_train)
```

```python
x_test=scalar.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={
    'criterion':['entropy','gini'],
    'max_depth':[None,10,20,30],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4]
}
```

```python
tree=DecisionTreeClassifier()
```

```python
grid_search=GridSearchCV(tree,param_grid=param_grid,cv=10,scoring='accuracy',verbose=1,n_jobs=-1)
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_parameter=grid_search.best_params_
```

```python
print(best_parameter)
```

```python
model=grid_search.best_estimator_
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(accuracy_score(y_pred,y_test))
```

```python
print(classification_report(y_pred,y_test))
```
