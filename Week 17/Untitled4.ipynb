```python
import pandas as pd
```

```python
df=pd.read_csv('loan_data.csv')
```

```python
df.head()
```

```python
num_cols=df.select_dtypes(include=['float64','int64']).columns.tolist()
```

```python
num_cols
```

```python
Q1=df[num_cols].quantile(0.25)
```

```python
Q3=df[num_cols].quantile(0.75)
```

```python
IQR=Q3-Q1
```

```python
outliers=((df[num_cols]<(Q1-1.5*IQR))|(df[num_cols]>(Q3+1.5*IQR))).any(axis=1)
```

```python
df_clean=df[~outliers]
```

```python
df_clean.shape
```

```python
df.shape
```

```python
df_clean
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df_clean['purpose']=le.fit_transform(df_clean['purpose'])
```

```python
df_clean
```

```python
df_clean['purpose'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df_clean.drop(['not.fully.paid'],axis=1)
```

```python
y=df_clean['not.fully.paid']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
x_train=scalar.fit_transform(x_train)
```

```python
x_test=scalar.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree=DecisionTreeClassifier()
```

```python
param_grid={
    'criterion':['entropy','gini'],
    'max_depth':[None,10,20,30],
    'min_samples_split':[15,5,10],
    'min_samples_leaf':[1,2,4]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search=GridSearchCV(tree,param_grid=param_grid,cv=5,scoring='recall')
```

```python
grid_search.fit(x_train,y_train)
```

```python
print(grid_search.best_params_)
```

```python
print(grid_search.best_estimator_)
```

```python
model=grid_search.best_estimator_
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_pred,y_test))
```

```python
data=pd.DataFrame(grid_search.cv_results_)
```

```python
data
```
