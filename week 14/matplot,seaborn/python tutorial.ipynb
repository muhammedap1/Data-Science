```python
class numbers:
    def __init__(self,start,end):
        self.start=start
        self.end=end
        self.current=start-1
    def __iter__(self):
        return self
    def __next__(self):
        if self.current < self.end:
            self.current+=1
            return self.current
        else:
            raise StopIteration

for i in numbers(5,10):
    print(i)
```

```python

def mean(values):
    return sum(values)/len(values)
def covariance(x,y):
    n=len(x)
    mean_x=mean(x)
    mean_y=mean(y)
    return sum((x[i]-mean_x) * (y[i]-mean_y) for i in range(n))/(n-1)
def correlation(x,y):
    cov=covariance(x,y)
    std_x=(sum((xi-mean(x))**2 for xi in x)/(len(x)-1))**0.5
    std_y=(sum((yi-mean(y))**2 for yi in y)/(len(y)-1))**0.5
    return cov/(std_x*std_y)
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

print("Covariance:", covariance(x, y))
print("Correlation:", correlation(x, y))
```
